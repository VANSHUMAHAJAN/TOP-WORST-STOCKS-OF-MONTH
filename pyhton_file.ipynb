RENAMING THE COLUMNS BY REMOVING THE UN-NECESSARY SPACES WITHIN THEM AND CAPITALIZING THEM

```python
new_columns=pd.DataFrame(list(df[['SYMBOL',' OPEN_PRICE',' CLOSE_PRICE','Month', 'Day', 'YEAR']].columns))[0].str.upper().str.replace(' ','')
```

FINDING THE UNIQUE VALUES IN YEARS,MONTHS,DAYS

```python
years=list(new_df['YEAR'].unique())
```

```python
months=list(new_df['MONTH'].unique())
```

```python
days=list(new_df['DAY'].unique())
```

CONVERTING YEAR TO INT FORMAT FROM OBJECTS

```python
new_df['YEAR']=df['YEAR']
```

DICTIONARY CONTAINING MONTHS NAME

```python
dict_of_months={1:'JAN',2:'FEB',3:'MAR',4:'APR',5:'MAY',6:'JUN',7:'JULY',8:'AUG',9:'SEPT',10:'OCT',11:'NOV',12:'DEC'}
```

```python
def filtering_companies(name,first,last,month,year):         ##FILTERING COMPANIES WITH INCOMPLETE INFORMATION
  a=new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]
  company_start_day=a[a['SYMBOL']==name]['DAY'].iloc[-1]
  company_end_day=a[a['SYMBOL']==name]['DAY'].iloc[0]

  value=int(company_start_day==first)+int(company_end_day==last)
  return value

  
```

```python
def performence(name,month,year):         ##CALCULATING PERFORMENCE OF THE COMPANIES IN %
  a=new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]
  closing_price=a[a['SYMBOL']==name]['CLOSE_PRICE'].iloc[0]
  opening_price=a[a['SYMBOL']==name]['OPEN_PRICE'].iloc[-1]

  performence=((closing_price-opening_price)/opening_price)*100
  return performence


```

```python
final_dict={}
for year in years:
  print(year)
  month_dict={}
  for month in months:                        ## AS 2022 HAS ONLY DATA OF 2 MONTHS ,i.e JAN,FEB
    if year==2022:
      if month not in [1,2]:
        continue
      
    
    print(month)
    ##NAME OF ALL THE COMPANIES WHOSE DATA IS AVAILABLE FOR THAT MONTH OF THAT YEAR
    list_of_companies=list(new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]['SYMBOL'].unique())
    #new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]
    
    ##FIRST DAY OF THE MONTH FOR WHICH DATA IS AVAILABLE
    first_day=new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]['DAY'].iloc[-1]
    ##LAST DAY OF THE MONTH FOR WHICH DATA IS AVAILABLE
    last_day=new_df[(new_df['YEAR']==year) & (new_df['MONTH']==month)]['DAY'].iloc[0]

    my_dict={}
    for company in list_of_companies:
      ##FILTERING COMPANIES WITH INCOMPLETE INFORMATION , AND GETTING THE DATA OF THE COMPANIES FOR WHICH DATA IS AVAILABLE FOR FIRST AND LAST DAY OF THE MONTH
      value=filtering_companies(company,first_day,last_day,month,year)


      ##CALCULATING THEIR PERFORMENCE OVER THE MONTH
      if value==2:
        result=performence(company,month,year)
        #print(f'{company}:{result}')

        my_dict[company]=result          ##STRORING DATA OF THE COMPANY FOR THE MONTH IN THE DICTIONARY

      else:
        continue



      

    ##CONVERTING my_dict TO A DATAFRAME
    latest_df=pd.DataFrame.from_dict(my_dict,orient='index')
    latest_df=latest_df.sort_values(by=0)
    
    ##CONVERTING DATAFRAME TO DICTIONARY
    worst_performence_dict=latest_df.head(10).to_dict()    ##DATAFRAME OF TEN WORST PERFORMING COMPANIES
    best_performence_dict=latest_df.tail(10).to_dict()     ##DATAFRAME OF TEN BEST PERFORMING COMPANIES



    month_dict[dict_of_months[month]]={}
    month_dict[dict_of_months[month]]['WORST_PERFORMER']=worst_performence_dict[0]
    month_dict[dict_of_months[month]]['BEST_PERFORMER']=best_performence_dict[0]
    print(f'Best_performence:{best_performence_dict}')
    print(f'Worst_performence:{worst_performence_dict}')

    print('\n')

    

  ## STORING ALL THE DATA IN THE FINAL DICTIONARY  
  final_dict[year]=month_dict

    

    




    
  
```

FINAL DICTIONARY CONTAING ALL THE RESULTS OF EVERY YEAR, EVERY MONTH
